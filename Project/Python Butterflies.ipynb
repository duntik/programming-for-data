```python
#Setting up

##required packages

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy import stats

df=pd.read_csv("Butterflydata.csv")

##Get overview
display(df)
df.info()
dfClean = df.copy()
print('Copied DataFrame')
print(dfClean)
```

```python
##Removing unnided columns 
dfClean = dfClean.drop(['Region', 'Experience', 'Concerns', 'Motivation', 'NBN', 'Temp', 'Wind'], axis=1)
##Clean DataFrame
print('\n\nDataFrame after deleting column\n--------------')
print(dfClean)
```

```python
# Cleaning

##Check for NA and remove if any
print(dfClean.isnull().values.any())
```

```python
##Check for outliers
Butts= dfClean['Butts']
Butts_box=plt.boxplot(Butts, positions=[2])

Butt_sp= dfClean['ButtSp']
Butt_sp_box=plt.boxplot(Butt_sp, positions=[3])

Flower= dfClean['Flowers']
Flower_box=plt.boxplot(Flower, positions=[4])

Flower_sp= dfClean['FlowerSp']
Flowers_sp_box=plt.boxplot(Flower_sp, positions=[5])

Shelter= dfClean['Shelter']
Shelter_box=plt.boxplot(Shelter, positions=[6])

Semi_nat= dfClean['PCSemi_Nat']
Semi_nat_box=plt.boxplot(Semi_nat, positions=[7])

plt.figure(figsize=(5, 6))
```

```python
###Plot alone as scale is v. different

Flowers= dfClean['Flowers']
Flowers_box=plt.boxplot(Flowers)
```

```python
###Outliers in Butterflys, Flower sp, and Flower, removing these: 
def remove_outliers(df, column):
    q1= column.quantile(0.25)
    q2= column.quantile(0.75)
    IQR = q2 - q1
    upper_limit = (q2 + 1.5*IQR)
    lower_limit = (q1 - 1.5*IQR)
    df_normal = df[(column <= upper_limit) & (column >= lower_limit)]
    return df_normal

dfClean=remove_outliers(dfClean, dfClean['Butts'])
dfClean=remove_outliers(dfClean, dfClean['FlowerSp'])
dfClean=remove_outliers(dfClean, dfClean['Flowers'])

dfClean.info()
display(dfClean)
```

```python
dfClean2 = dfClean.copy()
print('Copied DataFrame')
print(dfClean2)

##Check for normality
sns.displot(data=dfClean2, x="Butts")
sns.displot(data=dfClean2, x="ButtSp")
sns.displot(data=dfClean2, x="Flowers")
sns.displot(data=dfClean2, x="FlowerSp")
sns.displot(data=dfClean2, x="Shelter")
sns.displot(data=dfClean2, x='PCSemi_Nat')

###Some skew, get values
dfClean2.agg(['skew', 'kurtosis']).transpose()
```

```python
###Butts needs log transform, use arcsinh as 0 values
dfClean2.insert(len(dfClean2.columns), 'Butts_log', np.arcsinh(dfClean2['Butts']))
sns.displot(data=dfClean2, x="Butts_log")
```

```python
#Visualisations cont. 
##Butts abundance
fig, axs = plt.subplots(2, 2)
fig.tight_layout()
axs[0,0].scatter(dfClean2['Flowers'], dfClean2['Butts_log'])
axs[1,0].scatter(dfClean2['FlowerSp'], dfClean2['Butts_log'])
axs[0,1].scatter(dfClean2['Shelter'], dfClean2['Butts_log'])
axs[1,1].scatter(dfClean2['PCSemi_Nat'], dfClean2['Butts_log'])
axs[0,0].title.set_text('Flowers')
axs[1,0].title.set_text('Flower Species')
axs[0,1].title.set_text('Shelter')
axs[1,1].title.set_text('% Natural Habitat')
fig.text(0.02, 0.5, 'Butts Abundance', va='center', rotation='vertical')
plt.show()
```

```python
##Butts Diversity
fig, axs = plt.subplots(2, 2)
fig.tight_layout()
axs[0,0].scatter(dfClean2['Flowers'], dfClean2['ButtSp'])
axs[1,0].scatter(dfClean2['FlowerSp'], dfClean2['ButtSp'])
axs[0,1].scatter(dfClean2['Shelter'], dfClean2['ButtSp'])
axs[1,1].scatter(dfClean2['PCSemi_Nat'], dfClean2['ButtSp'])
axs[0,0].title.set_text('Flowers')
axs[1,0].title.set_text('Flower Species')
axs[0,1].title.set_text('Shelter')
axs[1,1].title.set_text('% Natural Habitat')
fig.text(0.02, 0.5, '# Butts Species', va='center', rotation='vertical')
plt.show()
```
